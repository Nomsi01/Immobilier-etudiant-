```python
import pandas as pd
```

```python
df = pd.read_csv("Immobilier-etudiant-/df_with_surface_and_rent_new.csv", sep =',')
```

## Ajout de la colonne 'prix_par_m2_min'(qui correspond souvent au loyer d'un studio) 

```python
df['prix_par_m2_min'] = df['Min Loyer']/df['Min Surface'] # On crée une nouvelle variable prix_par_m2_min 
df['prix_par_m2_min'][783]
print(df.iloc[783, 18:28])
#df['infos'][783]
```

## Construction des graphiques 

```python
df2 = df.groupby(['regions']).mean().sort_values('prix_par_m2_min')
df2 = df2.loc[(df2['prix_par_m2_min']>1)] 
```

```python
df2['regions'] = df2.index #je crée une colonne qui correspond à l'index (la colonne regions)
```

```python
# Construction d'un diagramme en bar de df2 
```

```python
import seaborn as sns
```

```python
df2.columns
```

```python
sns.barplot(y='regions', x='prix_par_m2_min', data=df2, width= 0.2, orient="h") #OK 
```

## Ajout de la colonne 'prix_par_m2_max'(qui correspond le plus souvent à des logements en colocation (T2/T3/T4)) 

```python
df['prix_par_m2_max'] = df['Max Loyer']/df['Max Surface'] # On crée une nouvelle variable prix_par_m2_min 
df['prix_par_m2_max'][783]
print(df.iloc[783, 18:29])
#df['infos'][783]
```

```python
df.groupby(['regions']).mean().sort_values('prix_par_m2_max')
```

## Construction des graphiques 

```python
df_max = df.groupby(['regions']).mean().sort_values('prix_par_m2_max')
df_max = df_max.loc[(df_max['prix_par_m2_min']>1)] 
```

```python
df_max.columns
df_max['regions'] = df_max.index #je crée une colonne qui correspond à l'index (la colonne regions)
```

```python
sns.barplot(y='regions', x='prix_par_m2_max', data=df_max, width= 0.2, orient="h").set_title('prix_par_m2_max')#OK 
```

## Ajout de la colonne 'prix_par_m2_mean'(qui correspond à la moyenne des loyers des logements porposés) 

```python
df['prix_par_m2_mean'] = df['Mean Loyer']/df['Mean Surface'] # On crée une nouvelle variable prix_par_m2_mean
df['prix_par_m2_mean'][783]
print(df.iloc[783, 18:30])
#df['infos'][783]
```

```python
df_mean = df.groupby(['regions']).mean().sort_values('prix_par_m2_mean')
df_mean = df_mean.loc[(df_mean['prix_par_m2_mean']>1)] 
df_mean.columns
df_mean['regions'] = df_mean.index #je crée une colonne qui correspond à l'index (la colonne regions)
sns.barplot(y='regions', x='prix_par_m2_mean', data= df_mean, width= 0.2, orient="h").set_title('prix_par_m2_mean')#OK 
```

## Compter le nombre de residence Crous par région 

```python
df['comptage_résidence']= 1
df.head(5)
df_nb_res = df.groupby(['regions']).sum().sort_values('comptage_résidence', ascending = False)

df_nb_res['regions'] = df_nb_res.index #je crée une colonne qui correspond à l'index (la colonne regions)
sns.barplot(y='regions', x='comptage_résidence', data= df_nb_res, width= 0.2, orient="h").set_title('Nombre de résidence par régions')#OK 
```

## Nombre de loyer min dispo par région 

```python
df['comptage_loyer_min'] = 1
df_loyer_min = df.loc[(df['prix_par_m2_min']>1)] 
df_loyer_min  = df_loyer_min.groupby(['regions']).sum().sort_values('comptage_loyer_min', ascending = False)

df_loyer_min['regions'] = df_loyer_min.index #je crée une colonne qui correspond à l'index (la colonne regions)
sns.barplot(y='regions', x='comptage_loyer_min', data= df_loyer_min, width= 0.2, orient="h").set_title('Nombre de loyers_min dospo par région')#OK 
```
