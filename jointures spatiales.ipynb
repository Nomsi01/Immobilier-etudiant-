```python
import pandas as pd
import nltk
nltk.download('punkt')
import geopandas as gpd
from shapely import wkt 
```

```python
df_crous = pd.read_csv("df_with_rent_and_surface.csv", sep = ',')
```

```python
df_crous = df_crous[ (df_crous["regions"] != "La Réunion") &
                    (df_crous["regions"] != "Guadeloupe") &
                    (df_crous["regions"] != "Martinique") &
                    (df_crous["regions"] != "Guyane")]
```

```python
df_crous = df_crous.reset_index(drop = True)
```

```python
"""
On récupère les donnés géographiques stockées dans la colonne geocalisation 
qui présente d'abord la latitude puis la longitude. 
"""

Longitude = []
Latitude = []
for j in range(0,df_crous.shape[0]):
    geocalisation = df_crous["geocalisation"][j]
    geocalisation = geocalisation.replace(',', ' ')
    word =  nltk.word_tokenize(geocalisation, language='french')
    Latitude.append(float(word[0]))
    Longitude.append(float(word[1]))
df_crous['Longitude'] = Longitude
df_crous['Latitude'] = Latitude
```

```python
df_crous.drop(['geocalisation'], axis = 1, inplace = True)
```

```python
df_crous
```

```python
gdf_crous = gpd.GeoDataFrame(
    df_crous, geometry=gpd.points_from_xy(df_crous.Longitude, df_crous.Latitude))

gdf_crous
```

```python
gdf_prive = pd.read_csv("loyer_privé.csv") #on reimporte la carte des loyers privés 
```

```python
gdf_prive['geometry'] = gdf_prive['geometry'].apply(wkt.loads) #on transforme en format wkt (sinon ça ne marche pas)
gdf_prive 
```

```python
gdf_prive = gpd.GeoDataFrame(gdf_prive)  #on transforme en geodataframe 
```

```python
joined_df = gpd.sjoin(gdf_crous, gdf_prive, how='left', op='within') #on réalise la jointure à gauche avec within 
#i.e si le point est dans le polynôme. 
```

```python
joined_df.to_csv(r'Loyer-joined-cities.csv', index = False) #pour stocker le nouveau dataframe 
```
