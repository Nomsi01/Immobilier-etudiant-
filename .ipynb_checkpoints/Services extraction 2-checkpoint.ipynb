```python
def tous_les_services (colonne):
    Services=[]
    for i in range (0, len(colonne)):
        try :
            a = re.split(' \r\n \xa0\xa0|\xa0\xa0|\xa0| \r\n \xa0|\n|\r',colonne[i]) #split permet de séparer les chaînes de caractères à chaque occurrence de \r\n, \xa0 etc
            for j in range (0,len(a)):
                Services.append(a[j])
        except:
            a= repr(colonne[i]) #repr permet de convertir la chaîne de caractère en string
            a = re.split(' \r\n \xa0\xa0|\xa0\xa0|\xa0| \r\n \xa0|\n|\r',a)
            for j in range (0,len(a)):
                Services.append(a[j])
    return Services

```

```python
def services_residence1 (residence):
    try :
        Services = re.split(' \r\n \xa0\xa0|\xa0\xa0|\xa0| \r\n \xa0|\n|\r',residence) #split permet de séparer les chaînes de caractères à chaque occurrence de \r\n, \xa0 etc
    except:
        residence = repr(residence) #repr permet de convertir la chaîne de caractère en string
        Services = re.split(' \r\n \xa0\xa0|\xa0\xa0|\xa0| \r\n \xa0|\n|\r',residence)
    return Services

```
